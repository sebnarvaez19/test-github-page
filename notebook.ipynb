Import dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Test Notebook

## Define Variables

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
```

## Plot data

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sine Wave')
ax.set_xlabel('x (radians)')
ax.set_ylabel('sin(x)')
plt.grid()
plt.show()
```
